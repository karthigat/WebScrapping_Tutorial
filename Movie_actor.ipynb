```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Mohan_(actor)"
```

```python
data  = requests.get(url).text 
```

```python
soup = BeautifulSoup(data,"html.parser")  # create a soup object using the variable 'data'
```

```python
table = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
mohan_data = pd.DataFrame(columns=["Movie", "Year", "Role", "Langugae", "Link"])

for row in table[3].tbody.find_all(name=["tr"]):
    col = row.find_all("td")
    if (col != []):
        year = col[0].text
        movie = col[1].text
        get_link = col[1].find('a', href=True)
        if get_link is not None:
            link = get_link.get('href')
        role = col[2].text.strip()
        language = col[3].text.strip()
        mohan_data = mohan_data.append({"Movie":movie, "Year":year, "Role":role, "Langugae":language, "Link":link}, ignore_index=True)
```
